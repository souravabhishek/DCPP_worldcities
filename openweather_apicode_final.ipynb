```python
import pandas as pd
import requests
from lxml import html
import openpyxl
import os
from datetime import datetime

user_api = 'efc67bce725eecbff01f384fc5fc392d'

```

```python
ps = openpyxl.load_workbook('worldcities_3.xlsx')
sheet = ps['Sheet1']
sheet
weather = {'city':[],'temp_city':[] ,'weather_desc':[], 'hmdt':[], 'wind_spd':[],'date_time':[]}
dfcity=[]
df=[]
for row in range(2,1565):
    CITY = sheet['A' + str(row)].value
    complete_api_link = "https://api.openweathermap.org/data/2.5/weather?q="+CITY+"&appid="+user_api
    api_link = requests.get(complete_api_link)
    api_data = api_link.json()
    try:
        temp = api_data['main']['temp']
        desc = api_data['weather'][0]['description']
        humidity = api_data['main']['humidity']
        wind_speed = api_data['wind']['speed']
        date_time = datetime.now().strftime("%d %b %Y | %I:%M:%S %p")
    except:
        continue
        
    weather['city'].append(CITY)
    weather['temp_city'].append(temp)
    weather['weather_desc'].append(desc)
    weather['hmdt'].append(humidity)
    weather['wind_spd'].append(wind_speed)
    weather['date_time'].append(date_time)
```
